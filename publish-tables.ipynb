## Publish tables & views to Pixeltable Cloud

Bring your tables and multimodal data to Pixeltable Cloud so teammates can explore, query, and replicate your data from anywhere. Publishing keeps schema, inputs, and generated outputs (including media) in sync—making it easy to share insight-rich datasets with a single URI.

To publish your first table, complete these prerequisites: 

1. Create a Pixeltable Cloud account
2. Generate an API key
3. Install the Pixeltable SDK

```python
%pip install pixeltable
```

```python
import pixeltable as pxt
```

```python
# any table listed here can be published
pxt.list_tables()
```

```python
# publish the base table with BOTH trailers
pxt.publish(
    source='my_vids', # use the table name as a string or the table object
    destination_uri="pxt://pixeltable:demos/tucci/tucci_trailers",
    access='public'
)
```

```python
# publish view 1 of the audio chunks
pxt.publish(
    source='audio_chunks',
    destination_uri="pxt://pixeltable:demos/tucci/tucci_audio",
    access='public'
)
```

```python
# publish view 2 of the transcript chunks
pxt.publish(
    source='transcript_chunks',
    destination_uri="pxt://pixeltable:demos/tucci/tucci_text",
    access='public'
)
```

```python
# publish view 3 of the franes
pxt.publish(
    source='frames',
    destination_uri="pxt://pixeltable:demos/tucci/tucci_frames",
    access='public'
)
```
