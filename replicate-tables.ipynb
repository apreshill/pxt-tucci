## Replicate tables & views from Pixeltable Cloud

You can replicate shared public tables locally with the Pixeltable SDK. A Pixeltable Cloud account is not required for access.

```python
%pip install pixeltable
```

```python
import pixeltable as pxt
```

### Base table

```python
# the local path is on the machine
# the table object `trailers` is in the current session only
trailers = pxt.replicate(
    remote_uri='pxt://pixeltable:demos/tucci/tucci_trailers',
    local_path='tucci_trailers_replica'  # Your local table name
)
```

```python
# use the local path in a new session with get_table
st_trailers = pxt.get_table('tucci_trailers_replica')
```

### View 1

```python
audio_chunks = pxt.replicate(
    remote_uri='pxt://pixeltable:demos/tucci/tucci_audio',
    local_path='tucci_audio_replica'  # Your local table name
)
```
